```python
x = df['age']
y = df['Income(k)']
```

```python
import numpy as np
x_pred = np.array([22])

```

```python
from sklearn.neighbors import KNeighborsRegressor

knn = KNeighborsRegressor(n_neighbors=3)
```

```python
knn.fit(x.values.reshape(-1,1), y)
```

```python
predict_income = knn.predict(x_pred.reshape(1,-1))
predict_income
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(x.values.reshape(-1,1), y)
```

```python
predicted_income_lr = lr.predict(x_pred.reshape(1,-1))
predicted_income_lr
```

```python
df_pred = pd.DataFrame(x_pred, columns=['age'])
df_pred
```

```python
df_pred['Income(k)'] = predict_income
df_pred
```

```python
df_pred.plot(kind='bar')
```

```python
df_pred = pd.DataFrame({'age': [22], 'Predicted_Income_KNN': predict_income, 'Predicted_Income_LR': predicted_income_lr})
print(df_pred)
```

```python
comparison_df = df.copy()
```

```python
comparison_df.loc[len(comparison_df.index)] = [22, np.nan]
```

```python
comparison_df = pd.concat([comparison_df, df_pred], axis=1)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['age'], df['Income(k)'], color='red', marker='*', label='Actual Data')
plt.scatter([22], predict_income, color='green', marker='o', label='Predicted Income (KNN)')
plt.scatter([22], predicted_income_lr, color='blue', marker='x', label='Predicted Income (Linear Regression)')
plt.xlabel('Age')
plt.ylabel('Income')
plt.title('Age vs. Income with Predictions')
plt.legend()
plt.show()
```
